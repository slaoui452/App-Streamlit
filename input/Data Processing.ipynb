```python
from sklearn.preprocessing import MinMaxScaler
```

```python
df_pca=df.drop(['Class','Time','Amount',],axis=1)
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df_pca.values)
```

```python
scaled_df = pd.DataFrame(scaled_data)
```

```python
scaled_df
```

```python
from sklearn.decomposition import PCA
```

```python
sklearn_pca = PCA(n_components=2)
```

```python
reduced_data = sklearn_pca.fit_transform(scaled_df.values)#ou sklearn_pca.fit_transform(scaled_data)
```

```python
reduced_data
```

```python
plt.figure(figsize=(8,6))
plt.scatter(reduced_data[:,0],reduced_data[:,1],c=df['Class'])
plt.xlabel('First principle component')
plt.ylabel('Second principle component')
```
